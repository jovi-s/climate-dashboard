```python
import matplotlib.pyplot as plt
import pandas as pd

def visualize_emissions(file_path):
    data = pd.read_excel(file_path)
    # Convert 'Year' column to numeric
    data['Year'] = pd.to_numeric(data['Year'], errors='coerce')

    # Plot 1: Total Carbon Emissions Over Time
    plt.figure(figsize=(10, 6))
    plt.plot(data['Year'], data['Total carbon emissions from fossil fuel consumption and cement production (million metric tons of C)'], label='Total Carbon Emissions')
    plt.title('Total Carbon Emissions Over Time')
    plt.xlabel('Year')
    plt.ylabel('Emissions (Million Metric Tons of C)')
    plt.grid(True)
    plt.legend()
    plt.show()

    # Plot 2: Breakdown by Source
    plt.figure(figsize=(10, 6))
    plt.stackplot(data['Year'], 
                  data['Carbon emissions from solid fuel consumption'],
                  data['Carbon emissions from liquid fuel consumption'],
                  data['Carbon emissions from gas fuel consumption'],
                  data['Carbon emissions from cement production'],
                  labels=['Solid Fuel', 'Liquid Fuel', 'Gas Fuel', 'Cement Production'])
    plt.title('Carbon Emissions Breakdown by Source')
    plt.xlabel('Year')
    plt.ylabel('Emissions (Million Metric Tons of C)')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.show()

    # Plot 3: Per Capita Emissions (after 1949)
    data_after_1949 = data.dropna(subset=['Per capita carbon emissions (metric tons of carbon; after 1949 only)'])
    
    plt.figure(figsize=(10, 6))
    plt.plot(data_after_1949['Year'], data_after_1949['Per capita carbon emissions (metric tons of carbon; after 1949 only)'], label='Per Capita Carbon Emissions')
    plt.title('Per Capita Carbon Emissions Over Time (After 1949)')
    plt.xlabel('Year')
    plt.ylabel('Per Capita Emissions (Metric Tons of C)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Call the function to visualize the data
visualize_emissions("../data/global.1751_2020.xlsx")
```
